```python
import matplotlib.pyplot as plt
plt.title('plotting')
plt.plot([10, 20, 30, 40]) # plot 함수는 직선 또는 꺾은선 형태의 그래프를 그릴 때 사용할 수 있는 명령어이다.
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [12, 43, 25, 15])
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('money')
plt.plot([10, 20, 30, 40], label='asc') # 증가를 의미하는 asc 라벨 
plt.plot([40, 30, 20 ,10], label='desc') # 감소를 의미하는 desc 라벨
plt.legend() # 여기에 범례 위치 지정 가능. 예 plt.lengend(loc = 5)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('color')
plt.plot([10, 20, 30, 40], color = 'skyblue', label='skyblue') # 색표시
plt.plot([40, 30, 20 ,10], 'pink', label='pink') # 색 표시
plt.legend() # 여기에 범례 위치 지정 가능. 예 plt.lengend(loc = 5)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('lineStyle')
plt.plot([10, 20, 30, 40], color = 'r', linestyle='--', label='dashed') 
plt.plot([40, 30, 20 ,10], color = 'g', ls=':', label='dotted')  # linestyle 대신에 ls 라고 할 수도 있다.
plt.legend() # 여기에 범례 위치 지정 가능. 예 plt.lengend(loc = 5)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('marker')
plt.plot([10, 20, 30, 40], 'r.', label='circle') 
plt.plot([40, 30, 20 ,10], 'g^', label='triangle up')  
plt.legend() # 여기에 범례 위치 지정 가능. 예 plt.lengend(loc = 5)
plt.show()
```
