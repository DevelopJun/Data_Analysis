```python
result = []
my_birthday = []
f = open('seoul.csv')
data = csv.reader(f)
next(data)
for row in data: 
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            result.append(float(row[-1]))
        if row[0].split('-')[1] == '10' and row[0].split('-')[2] == '27':
            my_birthday.append((float(row[-1])))
            
plt.plot(result, 'hotpink')
plt.show()
plt.plot(my_birthday, 'skyblue')
plt.show()
```

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)
high = [] #     
low = [] #      

for row in data:
    if row[-1] != '' and row[-2] != '':
        if 1984 <= int(row[0].split('-')[0]): # 1984    
            if row[0].split('-')[1] == '10' and row[0].split('-')[2] == '27':
                high.append(float(row[-1])) #    high   
                low.append(float(row[-2])) #    low   

plt.rc('font', family='Malgun Gothic')
# plt.title("June's Birthday change")
plt.title("    ")
plt.plot(high, 'green', label="high temperature")
plt.plot(low, 'yellow', label='low temperature')
plt.legend()
plt.show()
```
