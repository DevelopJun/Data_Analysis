```python
result = []
my_birthday = []
f = open('seoul.csv')
data = csv.reader(f)
next(data)
for row in data: 
    if row[-1] != '':
        if row[0].split('-')[1] == '08':
            result.append(float(row[-1]))
        if row[0].split('-')[1] == '10' and row[0].split('-')[2] == '27':
            my_birthday.append((float(row[-1])))
            
plt.plot(result, 'hotpink')
plt.show()
plt.plot(my_birthday, 'skyblue')
plt.show()
```
